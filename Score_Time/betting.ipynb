```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
 
betting = pd.DataFrame()

def scrape_page(base_url, page_number):
    # Set up the Chrome WebDriver (make sure to have the ChromeDriver executable in your PATH)
    driver = webdriver.Chrome()

    # Construct the URL for the current page
    current_url = f'{base_url}/#/page/{page_number}'

    # Open the page using Selenium
    driver.get(current_url)

    # Wait for the content to be loaded (adjust the timeout as needed)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'truncate.participant-name')))

    # Get the page source after JavaScript execution
    page_source = driver.page_source

    # Parse the HTML content of the page
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all <p> elements with class "truncate participant-name"
    team_names = soup.find_all('p', class_='truncate participant-name')
    
    time.sleep(5)

    # Find all elements with the specified class for team strength
    team_strengths = soup.find_all(class_='flex-center border-black-main min-w-[60px] max-w-[60px] flex-col gap-1 border-l border-opacity-10')

    home_team = [team_names[i].string for i in range(0, len(team_names), 2)]
    away_team = [team_names[i].string for i in range(1, len(team_names), 2)]

    betting['home_team'] = home_team
    betting['away_team'] = away_team

    home_team_betting = [team_strengths[i].get_text() for i in range(0, len(team_strengths), 3)]
    away_team_betting = [team_strengths[i].get_text() for i in range(2, len(team_strengths), 3)]

    betting['home_team_betting'] = home_team_betting
    betting['away_team_betting'] = away_team_betting

    return betting

# Set the base URL
epl_22_23 = 'https://www.oddsportal.com/football/england/premier-league-2022-2023/results/'
epl_21_22 = 'https://www.oddsportal.com/football/england/premier-league-2021-2022/results/'
epl_20_21 = 'https://www.oddsportal.com/football/england/premier-league-2020-2021/results/'

# Create an empty DataFrame to store the results
final_df = pd.DataFrame()

# Iterate through pages 1 to 8
for page_number in range(1, 9):
    df_1 = scrape_page(epl_22_23, page_number)
    df_2 = scrape_page(epl_21_22, page_number)
    df_3 = scrape_page(epl_20_21, page_number)
    final_df = pd.concat([df_1, df_2, df_3], ignore_index=True)

# Display the final DataFrame
print(final_df)
```
